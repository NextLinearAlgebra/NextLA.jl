```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import math
from numpy.linalg import norm
```

```python
def summarize_csv(filename):
    df = pd.read_csv(filename)

    summary = {
        'file': filename,
        'conv_sp': df['conv_sp'].value_counts().to_dict(),
        'conv_mp': df['conv_mp'].value_counts().to_dict(),
        'niters_sp': {
            'counts': df['niters_sp'].value_counts().to_dict(),
            'range': (df['niters_sp'].min(), df['niters_sp'].max())
        },
        'niters_mp': {
            'counts': df['niters_mp'].value_counts().to_dict(),
            'range': (df['niters_mp'].min(), df['niters_mp'].max())
        },
        'time_sp': {
            'count': df['time_sp'].count(),
            'range': (df['time_sp'].min(), df['time_sp'].max())
        },
        'time_mp': {
            'count': df['time_mp'].count(),
            'range': (df['time_mp'].min(), df['time_mp'].max())
        }
    }

    return summary
```
